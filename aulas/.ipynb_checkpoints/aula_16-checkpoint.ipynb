## Limpeza de dados

```python
df = pd.read_csv('https://raw.githubusercontent.com/vnikoofard/python_online_2020/master/data/Pacientes_compilados.csv', encoding = 'latin1')
```

```python
df
```

```python
'hoje está fazendo calor'.split(' ')
```

```python
'13-abr'.split('-')
```

```python
'13-abr'.split('-')[1]
```

```python
'13-abr'.split('-')[1] == 'abr'
```

```python
def fix(x):
    if x.split('-')[1] == 'abr':
        x = x.split('-')[0] + '-apr'
    x = x + '-2020'
    return pd.to_datetime(x)
```

```python
fix('13-abr')
```

```python
'13' + '-apr'
```

```python
df.columns
```

```python
df['Data de inicio de sintomas'].apply(fix)
```

```python
df.head()
```

```python
df['Data de inicio de sintomas'] = df['Data de inicio de sintomas'].apply(fix)
```
