# O efeito de borboleta

```python
import numpy as np
from scipy.integrate import odeint
#import matplotlib . pyplot as plt

# dados iniciais
sigma = 10
rho = 28
beta = 8/3
```

```python
x, y, z = [1,2,3]
```

```python
y
```

```python
def derivatives(state , t ) :
    x ,y , z = state
    xdot = sigma *( y - x )
    ydot = x *( rho - z ) -y
    zdot = x * y - beta * z
    return [ xdot , ydot , zdot ]
```

```python
# Determinar o tempo
t = np.arange(0 , 200 , 0.05)
#valores iniciais
y0 = [0 , 0.0001 , 0]
```

```python
sols = odeint(derivatives , y0 , t )
```

# ode

```python
import numpy as np
from scipy.integrate import ode

# dados iniciais
sigma = 10
rho = 28
beta = 8/3
```

Preste atenÃ§Ã£o na ordem dos argumentos

```python
def derivatives(t, state) :
    x ,y , z = state
    xdot = sigma *( y - x )
    ydot = x *( rho - z ) -y
    zdot = x * y - beta * z
    return [ xdot , ydot , zdot ]
```

Definir um integrador

```python
s = ode(derivatives) 
```

```python
t0 = 0
r0 = [0 , 0.0001 , 0]
```

```python
s.set_integrator('lsoda', min_step=0.0001)
s.set_initial_value(r0,t0)
```

```python
s.t
```

```python
s.y
```

```python
s.integrate(s.t)
```

```python
s.integrate(s.t + 0.01)
```
